## Analysis
- Degree of Centrality
- Most Influential
- Most Important Connection
- Best Betweener
- Shortest Path

```python
#General Connnection
nx.degree(soc_graph)
```

```python
#Degree of Connection
nx.degree(soc_graph, 2)
```

```python
nx.degree(soc_graph, 59)
```

```python
nx.degree(soc_graph, 241)
```

# Most Influential
- nx.degree_centrality()

```python
nx.degree_centrality(soc_graph)
```

```python
sorted(nx.degree_centrality(soc_graph).values())
```

```python
most_influential = nx.degree_centrality(soc_graph)
```

```python
for w in sorted(most_influential, key = most_influential.get, reverse = True):
    print(w, most_influential[w], sep=' ---> ')
```

# Most Important Connection
- nx.eigenvector_centrality

```python
nx.eigenvector_centrality(soc_graph, max_iter=4000)
```

```python
most_important_link = nx.eigenvector_centrality(soc_graph, max_iter=4000)
```

```python
for w in sorted(most_important_link, key = most_important_link.get, reverse = True):
    print(w, most_important_link[w], sep=" --> ")
```

# what is the shortest connection between 101 and 94

```python
nx.shortest_path(soc_graph, 101, 94)
```

```python
nx.shortest_path(soc_graph, 25, 67)
```

```python
nx.shortest_path(soc_graph, 127, 142)
```

# Betweener Centrality
- Bridge/ Connect

```python
plt.figure(figsize = (20,40))
nx.draw_networkx(soc_graph)
plt.show()
```

```python
nx.betweenness_centrality(soc_graph)
```

```python
best_connector = nx.betweenness_centrality(soc_graph)
for w in sorted(best_connector, key = best_connector.get, reverse = True):
    print(w, best_connector[w], sep=' ---> ')
```

## Building a sub group 
- Retreive all the nodes connected to a given node within a Networkx Graph
  - Method 1 BFS
  - Method 2 nx.node_connected_component(G,n)

```python
group1 = nx.bfs_tree(soc_graph, 127)
```

```python
group2 = nx.bfs_tree(soc_graph, 15)
```

```python
nx.draw_networkx(group1)
```

```python
nx.draw_networkx(group2)
```

```python
nx.node_connected_component(soc_graph, 127)
```
